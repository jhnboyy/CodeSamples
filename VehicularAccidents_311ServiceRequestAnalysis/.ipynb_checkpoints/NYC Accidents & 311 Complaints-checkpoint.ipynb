```python
#Importing needed libraries
import requests
import pandas as pd
from datetime import datetime
import geopandas as gpd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import matplotlib.cm as cm
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.model_selection import train_test_split
from statsmodels.discrete.count_model import ZeroInflatedPoisson
```

### Data Acquisition

```python
#Pulling in NYC 311 Complain data from NY Open Data APi (limit 1000 rows per request)
start_year= 2019
end_year= 2024
base_url = "https://data.cityofnewyork.us/resource/erm2-nwe9.json"
all_data = []
offset = 0
done = False
start_date = f"{start_year}-01-01T00:00:00"
end_date = f"{end_year + 1}-01-01T00:00:00"

while not done:
    query = {
        "$where": (
            f"created_date >= '{start_date}' AND created_date < '{end_date}' "
            f"AND agency_name = 'Department of Transportation' "
            f"AND complaint_type like '%Condition%' "
            f"AND incident_zip IS NOT NULL"
        ),
        "$limit": 1000,
        "$offset": offset
    }
    print(offset)
    response = requests.get(base_url, params=query)
    data_chunk = response.json()
    if not data_chunk:
        done = True
    else:
        all_data.append(data_chunk)
        offset += 1000
```

```python
## Formatting into proper df
df_311 = pd.concat([pd.DataFrame(i) for i in all_data])
## Checking data w head
df_311.head()
```

```python
## Kicking out to CSV file for reading in without API call (takes a while)
# df_311.to_csv("NYC311_RoadComplaints.csv",index=False)
## For reading in without API call
df_311 = pd.read_csv("NYC311_RoadComplaints.csv")
```

```python
### Pulling in NYC vehicle accident data (reusing some vars from 311 pull.)
base_url = "https://data.cityofnewyork.us/resource/h9gi-nx95.json"
all_data = []
offset = 0
done = False
while not done:
    query = {
            "$where": (
                f"crash_date >= '{start_date}' AND crash_date < '{end_date}' "
                f"AND zip_code IS NOT NULL"
            ),
            "$limit": 1000,
            "$offset": offset
        }
    print(offset)
    response = requests.get(base_url, params=query)
    data_chunk = response.json()
    if not data_chunk:
        done = True
    else:
        all_data.append(data_chunk)
        offset += 1000
```

```python
## Formatting into proper df
df_crashes = pd.concat([pd.DataFrame(i) for i in all_data])
## Checking data w head
df_crashes.head()
```

```python
## Kicking out to CSV file for reading in without API call (takes a while)
# df_crashes.to_csv("NYCVehicleCrashes.csv",index=False)
## For reading in without API call
df_crashes = pd.read_csv("NYCVehicleCrashes.csv")
```

```python
## pulling in geometry shapefile info / ZCTA conversion crosswalk table
zcta = gpd.read_file("https://data.cityofnewyork.us/resource/pri4-ifjk.geojson")
zcta = zcta[zcta["label"].notna() & (zcta["label"] != "")]
zcta["zip"] = zcta["label"].str.split(",")
zcta = zcta.explode("zip")
zcta["zip"] = zcta["zip"].str.strip()

# ZIP to ZCTA crosswalk to enrich the 311 and crash data 
crosswalk_df = zcta[["modzcta", "zip"]].drop_duplicates()
crosswalk_df = crosswalk_df.rename(columns={"zip": "zip_code"})
crosswalk_df["zip_code"] = crosswalk_df["zip_code"].astype(str)
crosswalk_df.head()
```

```python
## Processing the 311 Raw Data 
print(df_311.shape)#1038708 rows
df_311_clean = df_311[~df_311["agency"].isnull()]
print(df_311_clean.shape) # No rows removed

#converting raw Date strings ot dates
df_311_clean['created_date'] = pd.to_datetime(df_311_clean['created_date'], errors='coerce')
df_311_clean['closed_date'] = pd.to_datetime(df_311_clean['closed_date'], errors='coerce')

# Dropping null value rows for closed dates 
df_311_clean = df_311_clean[df_311_clean['closed_date'].notna()]
print(df_311_clean.shape) # 1022348 remaining

## Probelematic / incorrect non null close date values
df_311_clean = df_311_clean[df_311_clean['closed_date'] >= df_311_clean['created_date']]
print(df_311_clean.shape) #985377  remaining]

## Creating custom columns for generalize year_month value and column for days a ticket is open / active.
df_311_clean['year_month'] = df_311_clean['created_date'].dt.to_period('M').astype(str)
df_311_clean['days_open'] = (df_311_clean['closed_date'] - df_311_clean['created_date']).dt.days

## Finally limiting to the columns we want for analysis (More granular columns not needed atm)
df_311_limited = df_311_clean[['unique_key', 'year_month', 'borough', 'community_board', 'incident_zip',
                   'address_type', 'days_open', 'status', 'complaint_type', 'descriptor', 'resolution_description']]
print(df_311_limited.head())
```

```python
## Aggregating by the new categories we made for counts. 
agg_311 = df_311_limited.groupby(['year_month', 'modzcta', 'complaint_group_custom', 'days_open_group']).agg(
    complaint_count=('unique_key', 'nunique'))
agg_311 = agg_311.reset_index()
agg_311.head()
```

```python
### Informal overdispersion potential check for the numeric columns. This wil influence which model we want to try. 
numeric_cols_df = agg_311_wide.select_dtypes(include='number')
print(numeric_cols_df.columns)
for col in numeric_cols_df.columns:
    mean = numeric_cols[col].mean()
    var = numeric_cols[col].var(ddof=1)
    ## Getting the ratio (dont want to div by zero)
    if mean == 0:
        print("mean is zero")
    else:
        ratio = var / mean
    print(col)
    print(f"Mean: {round(mean,2)}")
    print(f"Variance: {round(var,2)}")
    print(f"Ratio = {round(ratio,2)}")
    
    if ratio > 1:
        print("potential overdispersed")
    elif ratio < 1:
        print("potential underdispersed")
    else:
        print("equals 1")
```

```python
## Overall Looks good. nulls not an issue and we have proper counts. Zero heavy, will need to deal with later on. 
## There is over dispersion in raw data, will need to look again later and consider with predictor vars. (in raw form)
```

```python
## The data seems heavily zero inflated, its count data so always positive, and also heavily right skewed.
## Attempting to normalized these variables some what.
## Ideal regression should be Zero Inflated Poisson Regression, provided there is no over dispersion.
```

```python
cols_to_transform = [
    "accident_count",
    "light_signal_based_complaint_open_less_than_1_week",
    "road_infra_based_complaint_open_less_than_1_month",
    "road_infra_based_complaint_open_less_than_1_week",
    "light_signal_based_complaint_open_less_than_1_month",
    "light_signal_based_complaint_open_a_month_or_more",
    "road_infra_based_complaint_open_a_month_or_more"]

# log(x + 1)
semifinal_data_log = semifinal_data_pruned.copy()
semifinal_data_log[cols_to_transform] = semifinal_data_log[cols_to_transform].applymap(lambda x: np.log1p(x))
```

```python
# Histograms of logged vars
numeric_cols = semifinal_data_log.select_dtypes(include=np.number).columns
semifinal_data_log[numeric_cols].hist(bins=30, figsize=(18, 20), layout=(4, 3))
plt.suptitle("Histograms of Log(x+1) Variables (No NZV Predicotrs)")
plt.show()
```

```python
#sqrt(x + 1)
semifinal_data_sqrt = semifinal_data_pruned.copy()
semifinal_data_sqrt[cols_to_transform] = semifinal_data_sqrt[cols_to_transform].applymap(lambda x: np.sqrt(x + 1))
```

```python
# Histograms of sq rt vars
numeric_cols = semifinal_data_sqrt.select_dtypes(include=np.number).columns
semifinal_data_sqrt[numeric_cols].hist(bins=30, figsize=(18, 20), layout=(4, 3))
plt.suptitle("Histograms of Sq Root Variables (No NZV Predictors)")
plt.show()
```

```python
# Apply cube root of (x + 1)
semifinal_data_cuberoot = semifinal_data_pruned.copy()
semifinal_data_cuberoot[cols_to_transform] = semifinal_data_cuberoot[cols_to_transform].applymap(lambda x: (x + 1)**(1/3))
```

```python
# Histograms of sq rt vars
numeric_cols = semifinal_data_cuberoot.select_dtypes(include=np.number).columns
semifinal_data_cuberoot[numeric_cols].hist(bins=30, figsize=(18, 20), layout=(4, 3))
plt.suptitle("Histograms of Cube Root (x+1) Variables (No NZV Predictors)")
plt.show()
```

```python
### I think the log and cube root transform both do a decent job of normalizing the dist. of the variables. Continuign with 
### these to transformations. 

## For picking a model we have zero heavy count data. Always positive with potential overdispersion (informal check earlier). 
## TO test for overdispersion we need to use a poisson model, which mandates integer-based target variable. 
## Trasnforming the data, but this time just the predictors. 
```

```python
predictors = [i for i in cols_to_transform if i != "accident_count"]
# log(x + 1)
semifinal_data_log = semifinal_data_pruned.copy()
semifinal_data_log[predictors] = semifinal_data_log[predictors].applymap(lambda x: np.log1p(x))
numeric_cols = semifinal_data_log.select_dtypes(include=np.number).columns
semifinal_data_log[numeric_cols].hist(bins=30, figsize=(18, 20), layout=(4, 3))
plt.suptitle("Histograms of Log(x+1) Variables (Only Predictors)")
plt.show()

# Cube Root (x+1)
semifinal_data_cuberoot = semifinal_data_pruned.copy()
semifinal_data_cuberoot[predictors] = semifinal_data_cuberoot[predictors].applymap(lambda x: (x + 1)**(1/3))
numeric_cols = semifinal_data_cuberoot.select_dtypes(include=np.number).columns
semifinal_data_cuberoot[numeric_cols].hist(bins=30, figsize=(18, 20), layout=(4, 3))
plt.suptitle("Histograms of Cube Root (x+1) Variables (Only Predictors)")
plt.show()
```

#### Overdispersion Test

```python
### Using Poisson MOdels to check for over dispersion before formally selecting models. 
## Poisson With Log
poisson_model_log = smf.glm(formula= "accident_count ~ " + " + ".join(predictors), 
                        data=semifinal_data_log, family=sm.families.Poisson()).fit()
observed_vals_log = semifinal_data_log["accident_count"]
fitted_vals_log = poisson_model_log.fittedvalues
pearson_resid_log = (observed_vals_log - fitted_vals_log) / np.sqrt(fitted_vals_log)
dispersion_log = np.sum(pearson_resid_log**2) / poisson_model_log.df_resid
print("Dispersion Estimate:", round(dispersion_log, 3))
```

```python
## Poisson with CubeRoot
poisson_model_cbrt = smf.glm(formula= "accident_count ~ " + " + ".join(predictors), 
                        data=semifinal_data_cuberoot, family=sm.families.Poisson()).fit()
observed_vals_cbrt = semifinal_data_cuberoot["accident_count"]
fitted_vals_cbrt = poisson_model_cbrt.fittedvalues
pearson_resid_cbrt = (observed_vals_cbrt - fitted_vals_cbrt) / np.sqrt(fitted_vals_cbrt)
dispersion_cbrt = np.sum(pearson_resid_cbrt**2) / poisson_model_cbrt.df_resid
print("Dispersion Estimate:", round(dispersion_cbrt, 3))
```

```python
### Both of these are indicative of over dipersion, so trying a negative binomial regression to accomodate for this reality,
## and Zero inflated poisson regression in order to adjust for the large amount of zeros in the data (structural vs random). 
```

#### Splitting the data into Training and Test Sets 

```python
### Splitting the data to training and test sets (Log)
y = semifinal_data_log["accident_count"]
x = semifinal_data_log[predictors] 
x_train_log, x_test_log, y_train_log, y_test_log = train_test_split(x, y, test_size=0.2, random_state=100)

### Splitting the data to training and test sets (Cube root)
y = semifinal_data_cuberoot["accident_count"]
x = semifinal_data_cuberoot[predictors] 
x_train_cbrt, x_test_cbrt, y_train_cbrt, y_test_cbrt = train_test_split(x, y, test_size=0.2, random_state=100)
```

```python
## AIC and BIC scores
print("AIC:",zip_model_log.aic)
print("BIC:",zip_model_log.bic)
x_test = sm.add_constant(x_test_log)
y_pred = zip_model_log.predict(x_test)
mae = mean_absolute_error(y_test_log, y_pred)
rmse = np.sqrt(mean_squared_error(y_test_log, y_pred))
print("MAE:", round(mae, 4))
print("RMSE:", round(rmse, 4))
```
